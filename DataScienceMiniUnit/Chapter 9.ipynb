```python
#rides = pd.read_sql_query("""select * from trip_data order by start_date limit 10""",'sqlite:///bikeshare.db', parse_dates=['start_date', 'end_date'])
rides = pd.read_sql_query("""select * from trip_data order by start_date""",'sqlite:///bikeshare.db', parse_dates=['start_date', 'end_date'])
rides['start_date'][0]
```

```python
rides.insert(4, "Duration", rides['end_date'] - rides['start_date'])
rides
```

```python
type(rides['Duration'][0])
```

```python
resamp = rides.set_index('start_date')
resamp = resamp.resample('D').mean()
resamp = resamp.reset_index()
Chart(resamp, width=600).mark_line().encode(
    x='start_date', y='duration').interactive()
```
